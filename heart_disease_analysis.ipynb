### 19. Plot continuous variables

```python
#In this question we are going to plot continuous variables.
# first we have to separate columns which contain continuous values and categorical values
```

```python
heartdata.columns
```

```python
# lets create an empty list
categ_val=[]
cont_val=[]
for column in heartdata.columns:
    if heartdata[column].nunique() <=10:
        categ_val.append(column)
    else: 
        cont_val.append(column)
```

```python
categ_val
```

```python
cont_val
```

```python
heartdata.hist(cont_val,figsize=(10,10))
plt.tight_layout()
plt.show()
```
